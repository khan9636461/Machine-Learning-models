# find the cost of j_function of linear regression in single parameters
# Using f(x)=w*x+b. For cost j(w,b)=1/2m(fw,b(x^i-y^1)^2) 

```python
x = [1500, 1700, 1800, 2000, 2100]  
y = [200000, 210000, 220000, 240000, 250000] 
b = 0  
w = 0  
m = len(x) 

# Hypothesis function: f_w(x) = b + w * x
def hypothesis(b, w, x_i):
    return b + w * x_i
def cost_function(b, w, x, y, m):
    cost = 0
    for i in range(m):
        prediction = hypothesis(b, w, x[i])
        cost += (prediction - y[i]) ** 2
    return (1 / (2 * m)) * cost
initial_cost = cost_function(b, w, x, y, m)
print(f"Initial Cost J(w,b): {initial_cost}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cost_function(theta):
    return theta**2

def gradient(theta):
    return 2 * theta

alpha = 0.1  
theta_left = -10.0  
iterations = 30  
theta_right = 10.0  
theta_values_left = [theta_left]
cost_values_left = [cost_function(theta_left)]
theta_values_right = [theta_right]
cost_values_right = [cost_function(theta_right)]

for i in range(iterations):
    grad = gradient(theta_left)
    theta_left = theta_left - alpha * grad  # Update theta
    theta_values_left.append(theta_left)
    cost_values_left.append(cost_function(theta_left))


for i in range(iterations):
    grad = gradient(theta_right)
    theta_right = theta_right - alpha * grad  # Update theta
    theta_values_right.append(theta_right)
    cost_values_right.append(cost_function(theta_right))

theta_range = np.linspace(-12, 12, 100)
cost_range = cost_function(theta_range)

plt.plot(theta_range, cost_range, label='Cost Function: J(theta) = theta^2')
plt.scatter(theta_values_left, cost_values_left, color='blue', label='Left-to-Right Path')
plt.plot(theta_values_left, cost_values_left, color='blue')
plt.scatter(theta_values_right, cost_values_right, color='red', label='Right-to-Left Path')
plt.plot(theta_values_right, cost_values_right, color='red')
plt.title('Bidirectional Gradient Descent Visualization')
plt.xlabel('Theta')
plt.ylabel('Cost J(Theta)')
plt.legend()
plt.show()
```

```python
import time
# Define the custom dot product function using a loop
def my_dot(a, b):
    dot_product = 0
    for i in range(len(a)):
        dot_product += a[i] * b[i]
    return dot_product


np.random.seed(1)
a=np.random.rand(1000000)
b=np.random.rand(1000000)

tic=time.time() # capture start time
c=np.dot(a,b)
toc=time.time() # for end time

print(f"np.dot(a,b) = {c:.4f}")
print(f"vectorized version duration : {1000*(toc-tic):.4f} ms")

tic=time.time() 
c=my_dot(a,b)
toc=time.time() 

print(f"my_dot(a,b) = {c:.4f}")
print(f"loop version duration : {1000*(toc-tic):.4f} ms")

del(a);del(b) # removes these big array from memory
```

 # Multiple Linear Regression with Gradient Descent

```python
import numpy as np

def compute_cost(X, y, w):
    m = len(y)
    predictions = X.dot(w)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))
    return cost

def gradient_descent(X, y, w, alpha, num_iterations):
    m = len(y)
    cost_history = np.zeros(num_iterations)

    for i in range(num_iterations):
        predictions = X.dot(w)
        gradient = (1  / m) * X.T.dot(predictions - y)
        w = w - alpha * gradient
        cost_history[i] = compute_cost(X, y, w)
    return w, cost_history

if __name__ == "__main__":
    X = np.array([[1, 1], [1, 2], [1, 3], [1, 4], [1, 5]]) 
    y = np.array([1, 2, 3, 4, 5])
    w = np.zeros(X.shape[1])
    alpha = 0.01
    num_iterations = 1000
    w, cost_history = gradient_descent(X, y, w, alpha, num_iterations)
    print("Final weights:", w)
    print("Final cost:", cost_history[-1])
```

# Model Implementation

```python
class multipleRegression():
    def __init__(self, lr=0.01, epochs=1000):
        self.lr=lr
        self.epochs=epochs
        self.weights=None
        self.bias=None
        
    def fit(self, x, y):
        n_samples, n_features=x.shape
        self.weights=np.zeros(n_features)
        self.bias=0.0
        cost=[]
        it=[]
        
        for i in range(self.epochs):
            y_pred=np.dot(x, self.weights) + self.bias
            cost.append(np.sum((y_pred - y)**2))
            it.append(i+1)
            dw= 1 / (n_samples)*np.dot(x.T,(y_pred-y))
            db= 1 / (n_samples)*np.sum(y_pred-y)
            self.weights=self.weights-(self.lr*dw)
            self.bias=self.bias-(self.lr*db)
        plt.plot(it, cost, color='b')
        
    def predict(self,x):
        y_pred=np.dot(x, self.weights)+self.bias
        return y_pred
            
        
```
